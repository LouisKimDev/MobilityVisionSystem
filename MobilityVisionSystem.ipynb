```python
from google.colab.patches import cv2_imshow
import numpy as np
import cv2
import matplotlib.pyplot as plt
from google.colab import drive

# 영상 불러오기
origin_img = cv2.imread("/content/drive/MyDrive/images/parrot.png")

# BGR 채널 순서를 RGB 채널로 변경
RGB_img = cv2.cvtColor(origin_img, cv2.COLOR_BGR2RGB)

# R, G, B 채널로 분할하기
Red_img, Green_img, Blue_img = cv2.split(RGB_img)

# 그림을 화면에 출력
plt.subplot(1, 4, 1)

# 원영상 출력
plt.title("origin_img")
plt.imshow(RGB_img)

# R 채널만 출력
RGB_img[:,:,0] = Red_img
RGB_img[:,:,1] = Red_img
RGB_img[:,:,2] = Red_img
plt.subplot(1,4,2)
plt.title("Red Channel")
plt.imshow(RGB_img)

# G 채널만 출력
RGB_img[:,:,0] = Green_img
RGB_img[:,:,1] = Green_img
RGB_img[:,:,2] = Green_img
plt.subplot(1,4,3)
plt.title("Green Channel")
plt.imshow(RGB_img)

# B 채널만 출력
RGB_img[:,:,0] = Blue_img
RGB_img[:,:,1] = Blue_img
RGB_img[:,:,2] = Blue_img
plt.subplot(1,4,4)
plt.title("Blue Channel")
plt.imshow(RGB_img)
plt.show()
```
