Question 2. RGB 이미지를 구해서 아래의 작업을 수행하시오

```python
from google.colab.patches import cv2_imshow
import numpy as np
import cv2
import matplotlib.pyplot as plt
from google.colab import drive

# 영상 불러오기
origin_img = cv2.imread("/content/drive/MyDrive/images/parrot.png")

# BGR 채널 순서를 RGB 채널로 변경
RGB_img = cv2.cvtColor(origin_img, cv2.COLOR_BGR2RGB)

# R, G, B 채널로 분할하기
Red_img, Green_img, Blue_img = cv2.split(RGB_img)

# 그림을 화면에 출력
plt.figure(figsize=(15,7))
plt.subplot(2, 4, 1)

# 1) RGB 이미지 출력
plt.title("origin_img")
plt.imshow(RGB_img)

# 2) R-channel 채널만 출력
RGB_img[:,:,0] = Red_img
RGB_img[:,:,1] = Red_img
RGB_img[:,:,2] = Red_img
plt.subplot(2,4,2)
plt.title("Red Channel")
plt.imshow(RGB_img)

# 3) G-channel 채널만 출력
RGB_img[:,:,0] = Green_img
RGB_img[:,:,1] = Green_img
RGB_img[:,:,2] = Green_img
plt.subplot(2,4,3)
plt.title("Green Channel")
plt.imshow(RGB_img)

# 4) B-channel 채널만 출력
RGB_img[:,:,0] = Blue_img
RGB_img[:,:,1] = Blue_img
RGB_img[:,:,2] = Blue_img
plt.subplot(2,4,4)
plt.title("Blue Channel")
plt.imshow(RGB_img)

# 5) Gray scale 이미지로 변환
RGB_img = cv2.cvtColor(origin_img, cv2.COLOR_BGR2RGB)
GRAY_img = cv2.cvtColor(RGB_img, cv2.COLOR_RGB2GRAY)
plt.subplot(2,4,5)
plt.title("Gray scale")
plt.imshow(GRAY_img, cmap = 'gray')

# 6) Binary 이미지로 변환
ret, BIN_img = cv2.threshold(GRAY_img, 128, 255, cv2.THRESH_BINARY)
plt.subplot(2,4,6)
plt.title("Binary")
plt.imshow(BIN_img, cmap = 'gray')


plt.show()
```

```python
# 7) 5번 작업을 수행한 후 출력 이미지에서 최대 화소값 출력
max = 0
for i in range(len(GRAY_img)):
  for j in range(len(GRAY_img)):
      if GRAY_img[i][j] > max:
        max = GRAY_img[i][j]

print(max)
print(np.max(GRAY_img))
```

```python
# 8) 5번 작업을 수행한 후 출력 이미지에서 최소 화소값 출력
min = 255
for i in range(len(GRAY_img)):
  for j in range(len(GRAY_img)):
      if GRAY_img[i][j] < min:
        min = GRAY_img[i][j]

print(min)
print(np.min(GRAY_img))
```

```python
# 9) 5번 작업을 수행한 후 전체 화소값의 평균값 출력
print(np.mean(GRAY_img))
```

Question 3. RGB 이미지 두 개(두 개의 이미지 사이즈가 동일해야 함)를 구해서 아래의 작업을 수행하시오.

```python
# 1) 두 이미지 모두 Gray scale로 변환

# 영상 불러오기
house1_org = cv2.imread("/content/drive/MyDrive/images/house.png")
house2_org = cv2.imread("/content/drive/MyDrive/images/house2.png")

# BGR 채널 순서를 RGB 채널로 변경
house1 = cv2.cvtColor(house1_org, cv2.COLOR_BGR2RGB)
house2 = cv2.cvtColor(house2_org, cv2.COLOR_BGR2RGB)

# Gray scale 이미지로 변환
house1_gr = cv2.cvtColor(house1, cv2.COLOR_RGB2GRAY)
house2_gr = cv2.cvtColor(house2, cv2.COLOR_RGB2GRAY)

# 2-a) 두 이미지를 Saturation 방법을 이용해 더하기
img_sat = cv2.add(house1_gr, house2_gr)

# 2-b) 두 이미지를 Wrapping 방법을 이용해 더하기
img_wrp = house1_gr + house2_gr

# 이미지 출력
plt.figure(figsize=(10,7))

plt.subplot(2,2,1)
plt.title("House1")
plt.imshow(house1_gr,cmap = 'gray')

plt.subplot(2,2,2)
plt.title("House2")
plt.imshow(house2_gr,cmap = 'gray')

plt.subplot(2,2,3)
plt.title("Saturation")
plt.imshow(img_sat,cmap = 'gray')

plt.subplot(2,2,4)
plt.title("Wrapping")
plt.imshow(img_wrp,cmap = 'gray')

plt.show()
```
